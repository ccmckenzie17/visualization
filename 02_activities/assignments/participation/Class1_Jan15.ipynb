```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
#set seed for reporducibility
np.random.seed(613)
#generate some data
x = np.arange(50)
y = np.random.randint(0, 100, 50)
```

```python
#pecify hight and width and create figure and axis objects
fig, ax = plt.subplots(figsize=(5, 3))
#create axis object as scatterplot of x and y
#ax.scatter(x,y)

#now run as bar
#ax.bar(x,y)

#or line
ax.plot(x,y)
```

```python
#activity, make histogram
fig, ax = plt.subplots(figsize=(5, 3))
ax.hist(y)
#including x in hist creates an error because histogram is only counts
```

```python
#make a plot with more detail
fig, ax = plt.subplots(figsize=(5, 3))
ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
fig.tight_layout()
ax.plot(x,y)
```

```python
#use font libraries to edit the label fonts
font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
ax.set_title('Total growth over time'
, fontdict = font1)
ax.set_ylabel('Total growth'
, fontdict = font2)
ax.set_xlabel('Years since start'
, fontdict = font2)
fig.tight_layout()
```

```python
#make plot with updated points and lines
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y,
           marker = '*',
           color = "indigo", 
           linestyle = '--',
           linewidth = 2) 
plt.show()
```

```python
#use hexcode to specify color
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y,
           marker = '*',
           color = "#7425b9", 
           linestyle = '--',
           linewidth = 2) 
plt.show()
```

```python
#add marker colors
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y,
           marker = '*',
           color = "#7425b9", 
           linestyle = '--',
           linewidth = 2,
           markeredgecolor = '#fa9359',
           markerfacecolor = '#000000',
           markersize = 12) 
plt.show()
```

```python
#gridline activity
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y,
           marker = '*',
           color = "#7425b9", 
           linestyle = '--',
           linewidth = 2,
           markeredgecolor = '#fa9359',
           markerfacecolor = '#000000',
           markersize = 12) 
ax.grid(axis = "y",
        color = "blue",
        linewidth = 2, 
        linestyle = '-.')
plt.show()
```

```python
#Activity: replicate graph
# libraries
import numpy as np
from numpy import linspace
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

# dataframe
df = pd.DataFrame({
'var1': np.random.normal(size=1000),
'var2': np.random.normal(loc=2, size=1000) * -1
})

# Fig size
plt.rcParams["figure.figsize"]=12,8

# plot density chart for var1
sns.kdeplot(data=df, x="var1",  fill=True, alpha=1)

# plot density chart for var2
kde = gaussian_kde(df.var2)
x_range = linspace(min(df.var2), max(df.var2), len(df.var2))

# multiply by -1 to reverse axis (mirror plot)
sns.lineplot(x=x_range*-1, y=kde(x_range) * -1, color='orange') 
plt.fill_between(x_range*-1, kde(x_range) * -1, color='orange')

# add axis names        
plt.xlabel("value of x")
plt.axhline(y=0, linestyle='-',linewidth=1, color='black')



# show the graph
plt.show()
```
